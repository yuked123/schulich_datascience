```python
# use loc function to find specific entries 

# loc function - get rows or columns with particular labels from the index 

# df.loc[index number] - output is all rows corresponding to that specific index number 

# in another words - you need to fill in the actual values or entry of the index for loc

df2.loc[1002]
```

```python
'''ArithmeticErrorlook for multiple index numbers - always use two brackets 
and put the two things side by side together within the same bracket 
'''

df2.loc[[1001, 1002]]
```

```python
df2.loc[df2['Product'] > 2]
```

```python
df2.loc[df2['Product'] > 2, 'Customers']
```

```python
# can also write as  - this becomes a dataframe format 
df2.loc[df2['Product'] > 2, ['Customers']]
```

```python
df2.loc[:1101, :'Province']

# BE AWARE - for loc - if use : operator - both start and end are included 
```

```python
# iloc function - get rows or columns at specific positions in the index 

# for iloc - we only need to fill in the order of the rows or index we want 

df2.iloc[0]

# df.iloc[the nth index]  -- in this case - the 0th index - the first row 
```

```python
df2.iloc[4]

# this means the 4+1 - 5th row 
```

```python
df2.iloc[2:4]

# this means from the 3rd row to the 4th row -- excluding the 5th row 
```
